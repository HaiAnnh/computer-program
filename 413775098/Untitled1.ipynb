```python
x = float(input("Enter the weight of the parcel(kg):"))
if x <= 5:
    print("The cost need to 50twd to send")
elif 5 < x <=10:
    print("The cost need to 70twd to send")
elif 10 < x <=15:
    print("The cost need to 90twd to send")
elif 15 <x <=20:
    print("The cost need to 110twd to send")
else:
    print("This was overweight, we can't send it")
```
