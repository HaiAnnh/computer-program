```python
#Find max in the list
def max_in_list(lst):
    if len(lst) ==1:
        return lst[0]
    sub_max = max_in_list(lst[1:])
    return lst[0] if lst [0] >sub_max else sub_max
lst = [int(x) for x in input("Enter number of list:").split()]
print(max_in_list(lst))
```

```python
#Kiem tra do an toan cua mat khau
def check_password(pw):
    has_digit = False
    has_alpha = False

    for ch in pw:
        if ch.isdigit():
            has_digit = True
        elif ch.isalpha():
            has_alpha = True
        length = len(pw)

        if(has_digit ^ has_alpha):
            if length < 6:
                return 'weak password'
            else:
                return 'seems safe password'
        else:
            if length < 6:
                return 'weak password'
            elif length >=10:
                return 'secure password'
            else:
                return 'normal password'
password = input('Enter your password')
result = check_password(password)
print(result)
```

```python
#
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)
n = int(input("Enter which item"))
print(fib(n))
```
